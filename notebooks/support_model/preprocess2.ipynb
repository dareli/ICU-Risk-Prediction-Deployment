```python
def audit_dataframe(dataframe):
    # summary table, ea row is a col
    summary = pd.DataFrame({
        "dtype": dataframe.dtypes.astype(str),
        # percentage of missing 
        "missing_pct": dataframe.isna().mean().round(3),
        #unique values?
        "n_unique_including_nan": dataframe.nunique(dropna=False),
    })

    # find the most common value and its frequency (as a %)
    summary["top_value_pct"] = [
        dataframe[col]                          
        .value_counts(dropna=False, normalize=True)  
        .iloc[0] # most frequent 
        .round(3)                               
        for col in dataframe.columns
    ]
    return summary.sort_values(["dtype", "n_unique_including_nan", "missing_pct"])


audit = audit_dataframe(df)
display(audit)


# checks

print("\nobject columns:")

display(
    df.select_dtypes(include="object")
      .nunique(dropna=False)
      .sort_values()
)


print("\nnumeric columns containing -1:")

# dictionary to track how many times -1 appears in the num cols
neg1_counts = {
    col: int((df[col] == -1).sum())  
    for col in df.select_dtypes(include="number").columns  
    if (df[col] == -1).any()  
}

display(pd.Series(neg1_counts).sort_values(ascending=False))
```

### **fix duplicate _x/_y columns first**
- in the merged file, there's several pairs are exact duplicates?

```python
df_clean = df.copy()

duplicate_pairs = [
    ("intaketotal_x", "intaketotal_y", "intaketotal"),
    ("outputtotal_x", "outputtotal_y", "outputtotal"),
    ("dialysistotal_x", "dialysistotal_y", "dialysistotal"),
    ("nettotal_x", "nettotal_y", "nettotal"),
    ("io_missing_x", "io_missing_y", "io_missing"),
    ("hospitalid_x", "hospitalid_y", "hospitalid"),
]

for left_col, right_col, new_name in duplicate_pairs:
    if left_col in df_clean.columns and right_col in df_clean.columns:
        same_values = (df_clean[left_col].fillna("__NA__") == df_clean[right_col].fillna("__NA__")).all()
        print(f"{left_col} vs {right_col} identical? -> {same_values}")
        if same_values:
            df_clean = df_clean.rename(columns={left_col: new_name})
            df_clean = df_clean.drop(columns=[right_col])
        else:
            print(f"warning!!: {left_col} and {right_col}  not identical. review again pls")

print("\nthe shape after duplicate cleanup:", df_clean.shape)
display(df_clean.head())
```

### **convert sentinel '-1' values to missing values**
- had to do some research but a lot of the clinical variables use '-1' as a placeholder for “not measured / not available ” info.  
- so not a real physiologic value, so we convert those to nan

```python

numeric_cols = df_clean.select_dtypes(include="number").columns.tolist()

neg1_before = {
    col: int((df_clean[col] == -1).sum())
    for col in numeric_cols
    if (df_clean[col] == -1).any()
}

for col in numeric_cols:
    if (df_clean[col] == -1).any():
        df_clean[col] = df_clean[col].replace(-1, np.nan)

neg1_after = {
    col: int((df_clean[col] == -1).sum())
    for col in df_clean.select_dtypes(include="number").columns
    if (df_clean[col] == -1).any()
}

print("columns that had -1:")
display(pd.Series(neg1_before).sort_values(ascending=False))

print("\nremaining cols still containing -1 (should be empty or very small if truly valid):")
display(pd.Series(neg1_after))
```

----------------------------------------------
### **target and drop leakage/bloat cols**

why these drops?
- **ids** are not useful predictors
- **discharge/death/ readmit outcome fields** directly reveal the target or happen after the prediction moment
- raw **`pasthistoryvalue`** is the main source of dummy explosion
- we already keep the cleaner grouped history flags like `hx_cardio`, `hx_renal`, etc.
- visit counters can leak future trajectory for readmission-style outcomes

```python
target_col = "bad_outcome"

drop_columns = [
    # ids
    "patientunitstayid",

    # raw history text/high-cardinality 
    "pasthistoryvalue",

    # post-outcome/leakage variables
    "hospitaldischargestatus",
    "unitdischargestatus",
    "readmit",
    "ICU_death",
    "hospital_death",
    "ICU_readmit",
    "readmit_clean",

    # likely downstream or trajectory-style counters
    "visitnumber",
    "unitvisitnumber",
]

existing_drop_columns = [col for col in drop_columns if col in df_clean.columns]

model_df = df_clean.drop(columns=existing_drop_columns)

print("dropped columns:")
print(existing_drop_columns)

print("\nshape after manual drops:", model_df.shape)
```

```python
## another missinginess audit

missing_counts = model_df.isnull().sum()
missing_percent = (missing_counts / len(model_df)) * 100

missing_summary = (
    pd.DataFrame({
        "missing_count": missing_counts,
        "missing_percent": missing_percent
    })
    .sort_values(by="missing_percent", ascending=False)
)

# show only columns with missing values
missing_summary = missing_summary[missing_summary["missing_count"] > 0]

print("missingness sorted:")
display(missing_summary)
```

```python
# colsw/ >30% missing
high_missing = missing_summary[missing_summary["missing_percent"] > 30]

print("columns with >30% missing:")
display(high_missing)
```
